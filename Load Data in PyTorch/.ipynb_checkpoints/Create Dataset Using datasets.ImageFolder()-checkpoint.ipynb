### Create Dataset Using datasets.ImageFolder()

```python
from torch.utils.data import Dataset, DataLoader
import requests
import zipfile
from pathlib import Path
import os

import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.utils import data
```

```python
!pwd
```

```python
# Change this for your own directory.
ROOT_DIR = Path("/root/jupyter_projects")  # Your working directory.
DATASET_DIR = ROOT_DIR / "Data"  # Folder to store dataset.

data_path = DATASET_DIR / "Caltech101"

# Contains 102 classes of image (101 objects + background)
```

```python
print(ROOT_DIR)
print(DATASET_DIR)
```

```python

# Uncomment this if you haven't downloaded the dataset.
url = "https://bucket-oranges.s3.us-west-2.amazonaws.com/Caltech101.zip"
file_name = "Caltech101.zip"
file_path = DATASET_DIR / file_name

r = requests.get(url, allow_redirects=True)

open(file_path, 'wb').write(r.content)
print('*Dataset created*')
```

```python

# Uncomment this if you haven't downloaded the dataset.
path_zip_file = file_path
directory_extract_to = DATASET_DIR

with zipfile.ZipFile(path_zip_file, 'r') as zip_ref:
    zip_ref.extractall(directory_extract_to)
print('*Successfully unzip file*')

path_zip_file.unlink()  # Delete the zip file.
```

```python
print(len(next(os.walk(data_path))[1]))
```

```python
img_transforms = transforms.Compose([
    transforms.CenterCrop(227),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]),
])
```

```python
dataset = datasets.ImageFolder(
    root=data_path,
    transform=img_transforms
)
print(f"Total number of samples: {len(dataset)}")
```

```python
print(type(dataset))
print(dataset[0][0].shape)  # X
print(dataset[0][1])  # y
```

```python
dataloader = data.DataLoader(dataset, batch_size=64)
```

```python
print(type(dataloader))
```
