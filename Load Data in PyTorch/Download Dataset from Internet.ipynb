```python
from pathlib import Path
import requests
import zipfile  # Unzip `.zip` file.
import tarfile  # Unzip `.tar` file.
```

```python
ROOT_DIR = Path("/root/jupyter_projects")  # Working directory.

DATASET_DIR = ROOT_DIR / "Data"  # Directory to store dataset.
DATA_DIR = DATASET_DIR / "Flowers102"  # Directory of dataset.

DATA_DIR.mkdir(parents=True, exist_ok=True)  # Create directory if it do not exist.
```

```python
URL = "https://bucket-oranges-1310562101.cos.ap-guangzhou.myqcloud.com/Flowers102.tgz"
file_name = "Flowers102.tgz"
file_path = DATA_DIR / file_name  # Download file path.

r = requests.get(URL, allow_redirects=True)
open(file_path, 'wb').write(r.content)

print('Dataset downloaded successfully!')
```

```python
zip_file_path = file_path
extract_directory = DATA_DIR

# For `.zip` file:
# with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
#     zip_ref.extractall(extract_directory)
    
# For `.tar.gz` file:
with tarfile.open(zip_file_path, "r") as tar_ref:
    tar_ref.extractall(extract_directory)
    
Path.unlink(file_path)  # Remove zip file.

print('Dataset extracted successfully!')
```
