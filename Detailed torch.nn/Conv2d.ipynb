## Convolution Layers

```python
import torch

torch.nn.Conv2d(
    in_channels,  # input channels
    out_channels,  # output channels
    kernel_size,
    stride=1, 
    padding=0, 
    dilation=1, 
    groups=1, 
    bias=True, 
    padding_mode='zeros', 
    device=None, 
    dtype=None
)

# Normal Convolution
# 3D Convolution
# Group Convolution    :    At groups=in_channels
# Depthwise Convolution:    At groups=in_channels
```

### Example1

```python
torch.nn.Conv2d(
    in_channels = 16,
    out_channels = 32,
    kernel_size = (3, 5),  # non-square kernels
    stride=(2, 1),  # unequal stride
    padding=(4, 2)  # with padding
)
```
