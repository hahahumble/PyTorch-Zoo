```python
# Change this for your own directory.
ROOT_DIR = Path("/home/studio-lab-user/sagemaker-studiolab-notebooks")  # Your working directory.
DATASET_DIR = ROOT_DIR / "Data"  # Folder to store dataset.

DATA_DIR = DATASET_DIR / "Caltech_Birds2011"

DATA_DIR.mkdir(parents=True, exist_ok=True)  # Create directory if it do not exist.
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')  # Define pytorch device.
```

```python
URL = "https://www.dropbox.com/s/i49ocwwkfjep0jg/caltech_birds2011.zip?raw=1"
file_name = "CUB_200_2011.zip"
file_path = DATA_DIR / file_name  # Download file path.
```

```python
print(file_path)
```

```python
r = requests.get(URL, allow_redirects=True)
open(file_path, 'wb').write(r.content)

print('Dataset downloaded successfully!')
```

```python
zip_file_path = file_path
extract_directory = DATA_DIR

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_directory)    

Path.unlink(file_path)  # Remove zip file.

print('Dataset extracted successfully!')
```

```python
image_dir = DATA_DIR / "caltech_birds2011/CUB_200_2011/images"
```

## Data Augmentation

```python
img_transforms = transforms.Compose([
    transforms.CenterCrop(227),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
])
```

```python
dataset = datasets.ImageFolder(
    root=image_dir,
    transform=img_transforms
)
```
