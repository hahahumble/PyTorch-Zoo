# ModuleList and ModuleDict

```python
import torch.nn as nn
import torch
```

## ModuleList

```python
class MyModule(nn.Module):
    def __init__(self, sizes):
        super().__init__()
        conv_blocks = [nn.Linear(in_f, out_f) for in_f, out_f in zip(sizes, sizes[1:])]
        self.layers = nn.ModuleList(conv_blocks)
        self.trace = []
        
    def forward(self,x):
        for layer in self.layers:
            x = layer(x)
            self.trace.append(x)
        return x
```

```python
model = MyModule([1, 16, 32])  # (1, 16), (16, 32)

model(torch.rand((4, 1)))  # (1, 1)

[print(trace.shape) for trace in model.trace]
```
