## Depthwise Separable Convolution

https://pytorch.org/docs/0.3.1/nn.html#torch.nn.Conv2d

### Depthwise Convolution

```python
import torch.nn as nn
from torch.nn import Conv2d
```

```python
class depthwise_conv(nn.Module):
    def __init__(self, in_channel, kernel_size):
        super().__init__()
        self.depthwise = nn.Conv2d(in_channel, in_channel, kernel_size=kernel_size, groups=in_channel)
        
    def forward(self, x):
        output = self.depthwise(x)
        return output
```

### Pointwise Convolution

```python
class pointwise_conv(nn.Module):
    def init(self, in_channel, out_channel):
        super(self).__init__()
        self.pointwise = nn.Conv2d(in_channel, out_channel, kernel_size=1)
        
    def forward(self, x):
        output = self.pointwise(x)
        return output
```

### Depthwise Separable Convolution

```python
class depthwise_separable_conv(nn.Module):
    def init(self, in_channel, out_channel, kernel_size):
        super(self).__init__()
        self.depthwise = nn.Conv2d(in_channel, in_channel, kernel_size=kernel_size, groups=in_channel)
        self.pointwise = nn.Conv2d(in_channel, out_channel, kernel_size=1)
        
    def forward(self, x):
        output = self.pointwise(x)
        return output
```
